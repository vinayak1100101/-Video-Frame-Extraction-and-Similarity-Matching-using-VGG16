```python
import os
import cv2
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from google.colab.patches import cv2_imshow

# Function to extract frames from the video
def extract_frames(video_path, interval=100):
    cap = cv2.VideoCapture(video_path)
    frames = []
    count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if count % interval == 0:
            frames.append(frame)

        count += 1
    cap.release()

    return frames

# Function to list all files in a directory and store them in a list
def list_files_in_directory(directory_path):
    files_list = []
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            if file.endswith(('.mp4', '.avi', '.mov')):  # Add other video formats if needed
                files_list.append(os.path.join(root, file))
    return files_list

# Example directory path in Colab
directory_path = "/content/videos"  # Replace with your directory path

# Get the list of all video files in the directory
files_list = list_files_in_directory(directory_path)
print(f"Found {len(files_list)} video files.")

# Extract frames from all videos
all_frames = []
for file_path in files_list:
    frames = extract_frames(file_path)
    print(f"Extracted {len(frames)} frames from {file_path}")
    all_frames.extend(frames)  # Flatten the list of lists

# Check if any frames were extracted
if not all_frames:
    print("No frames were extracted from the videos.")
    exit()

# Load the VGG16 model pre-trained on ImageNet
model = VGG16(weights='imagenet', include_top=False)

# Function to extract features using VGG16
def extract_features(img):
    img = cv2.resize(img, (224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    features = model.predict(img)
    return features.flatten()

# Load the query image
query_img_path = "/content/photos/ntrp_8.png"  # Ensure this path is correct
query_img = cv2.imread(query_img_path)
if query_img is None:
    print(f"Error: Could not load image at path {query_img_path}")
    exit()

# Extract features from the query image
query_features = extract_features(query_img)
print(f"Extracted features from query image.")

# Function to find the best matching frame
def find_best_match(query_features, frames):
    best_match = None
    best_similarity = -1
    for idx, frame in enumerate(frames):
        frame_features = extract_features(frame)
        similarity = cosine_similarity([query_features], [frame_features])[0][0]
        if similarity > best_similarity:
            best_similarity = similarity
            best_match = frame
        print(f"Frame {idx}: Similarity = {similarity}")
    return best_match

# Find the best matching frame
best_match = find_best_match(query_features, all_frames)

# Display the best matching frame (if needed)
if best_match is not None:
    cv2_imshow(best_match)
else:
    print("No matching frame found.")
```
